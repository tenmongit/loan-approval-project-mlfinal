```python
import pandas as pd

df = pd.read_csv('../data/raw/loan_data.csv')


df = df.dropna()

if df['loan_approved'].dtype == object:
    df['loan_approved'] = df['loan_approved'].map({
        'Yes': 1,
        'No': 0,
        'Y': 1,
        'N': 0
    })


df = df.drop(columns=['name'])

df.to_csv('../data/processed/cleaned.csv', index=False)
```
