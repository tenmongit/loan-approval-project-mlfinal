```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Load processed dataset
df = pd.read_csv('loan-approval-project-mlfinal\data\processed\cleaned.csv')

# Debugging: Print columns to identify the correct loan status column
print("DataFrame columns:", df.columns)

# Plot distribution of loan status
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='loan_approved')

plt.title('Loan Approval Distribution')
plt.xlabel('Loan Status')
plt.ylabel('Count')

# Create reports directory if it doesn't exist
os.makedirs('../reports', exist_ok=True)

# Save figure for report
plt.tight_layout()
plt.savefig('../reports/status_distribution.png')
plt.show()
```
